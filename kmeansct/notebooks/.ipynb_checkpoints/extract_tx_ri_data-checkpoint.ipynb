```python
import numpy as np
def tx_count_diff(x,y):
    count_diff = x-y
    if count_diff < 0:
        return 0.0
    else:
        return count_diff
```

```python
def tx_count_moving_average(col1,col2,col3,col4,col5,col6,col7,col8):
    min = 0
    increase = []
    for i in (col1,col2,col3,col4,col5,col6,col7,col8):
        increase.append(i-min)
        min = i
    average = np.mean(increase)
    #print(average)
    if average < 0:
        average = 0
    return round(average)
```

```python
url = r"C:\Python\pluralsight\covid-19-analytics\input_data\TX\raw\Texas COVID-19 Active Case Data by County.xlsx"
df = pd.read_excel(url, header=[0], skiprows=[0, 1])
tx_columns = df.columns
df = df.rename(columns={tx_columns[-1]: 'case_counts_today',tx_columns[-8]: 'case_counts_lastweek',
                            tx_columns[1]: 'city'})
dropindex = df[df['case_counts_today'].isna() == True].index
df.drop(index=dropindex, inplace=True)
df['average_counts'] = np.vectorize(tx_count_moving_average)(df['case_counts_lastweek'],
                                                                df[tx_columns[-7]],
                                                                df[tx_columns[-6]],
                                                                df[tx_columns[-5]],
                                                                df[tx_columns[-4]],
                                                                df[tx_columns[-3]],
                                                                df[tx_columns[-2]],
                                                                df['case_counts_today'])

df = df.loc[0:,['city','case_counts_lastweek','case_counts_today','average_counts']]
df.reset_index()
#print(df)
#df['case_counts_7day'] = np.vectorize(tx_count_diff)(df['case_counts_today'],df['case_counts_lastweek'])
print(df.info())
```

```python
pop = r'C:\Python\pluralsight\covid-19-analytics\input_data\lookup\TX_population.csv'
df_poulation = pd.read_csv(pop)
df_poulation = df_poulation.iloc[0:,[1,4]]
tx_columns = df_poulation.columns
df_poulation = df_poulation.rename(columns={tx_columns[0]: 'city',tx_columns[-1]: 'est_population'})
df_poulation['est_population_per_100k_factor'] = df_poulation['est_population'].apply(lambda x : 100000/x)
print(df_poulation.info())
```

```python
df = df.merge(df_poulation,how='inner',left_on='city',right_on='city')
df = df.loc[0:, ['city', 'average_counts', 'est_population_per_100k_factor']] ###
df.info()
```

```python
def deduce_counts(x,y):
    if y < 1:
        return x*y
    else:
        return x
```

```python
df['case_counts'] = np.vectorize(deduce_counts)(df['average_counts'],df['est_population_per_100k_factor'])
```
